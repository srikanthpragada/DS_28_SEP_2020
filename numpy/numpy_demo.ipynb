#### Masking

```python
a [(a > 40) | (a < 20)]  # ORing 
```

```python
a= np.random.randint(1,10,10)
a
```

```python
# Returns a list of booleans 
def prime(a):
    lst = []
    for n in a:
        for i in range(2,n//2 + 1):
            if n % i == 0 :
                lst.append(False)
                break
        else:
            lst.append(True)
        
    return lst  # List of booleans 
```

```python
a = np.random.randint(1,100,10)
print(prime(a))
a[prime(a)]
```
